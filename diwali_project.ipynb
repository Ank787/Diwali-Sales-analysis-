```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Exploratory Data Analysis

```python
diwali_df.columns
```

```python
at = sns.countplot(x= 'Gender',data = diwali_df)
```

```python
at = sns.countplot(x = 'Gender', data = diwali_df)

for bars in at.containers:
    at.bar_label(bars)
```

```python
at = sns.countplot(x = 'Marital_Status', data = diwali_df)

for bars in at.containers:
    at.bar_label(bars)
```

```python
diwali_df.groupby(['Gender'], as_index =False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
```

```python
sales_gen = diwali_df.groupby(['Gender'], as_index =False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sns.barplot(x ='Gender', y = 'Amount', data = sales_gen);
```

  Shopping done by female is greater than male

```python
diwali_df.columns
```

```python
mar_gen = diwali_df.groupby(['Marital_Status','Gender'], as_index =False)['Amount'].sum().sort_values(by = 'Amount',ascending= False)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(x ='Marital_Status', y = 'Amount', data = mar_gen, hue = 'Gender');
```

Shopping done by unmarried is greater than married



On the basis of Age

```python
sns.countplot(data = diwali_df, x= 'Age_Group');
```

hue is use to differenciate between male and female in Gender

```python
bt = sns.countplot(data= diwali_df, x = 'Age_Group', hue = 'Gender' );

for bars in bt.containers:
    bt.bar_label(bars)
```

Total amount vs Age Group

```python
sales_age = diwali_df.groupby(['Age_Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = 'Age_Group',y = 'Amount', data = sales_age);
```

Most of the buyers are age between 26-35 

State

```python
sales_state = diwali_df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(15)
#width is 30 and height 10
sns.set(rc={'figure.figsize':(30,10)})
sns.barplot(data = sales_state,x ='State', y = 'Orders');
```

According to graph its show that sales is greater in Uttar Pradesh then Maharashtra and so on

```python
diwali_df.columns
```

```python
sales_state = diwali_df.groupby(['State','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(15)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data = sales_state , x= 'State',y = 'Amount' , hue = 'Gender');
```

```python
sales_zone = diwali_df.groupby(['Zone','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = sales_zone , x = 'Zone', y = 'Amount', hue = 'Gender');
```

Central zone is spend more amount

```python
order_zone = diwali_df.groupby(['Zone','Gender'], as_index = False)['Orders'].sum().sort_values(by ='Orders', ascending = False).head(10)

sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data = order_zone, x='Zone', y= 'Orders' , hue = 'Gender');
```

# On the basis of Occupation 

```python
diwali_df.columns
```

```python
sales_occ = diwali_df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(20)

sns.set(rc = {'figure.figsize':(15,10)})
sns.barplot(data = sales_occ , x='Occupation' , y = 'Amount');
```

IT sector people spend more money on diwali

# On the basis of Product category

```python
sales_pro = diwali_df.groupby(['Product_Category'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False).head(15)

sns.set(rc= {'figure.figsize':(30,5)})
sns.barplot(data = sales_pro , x='Product_Category', y= 'Orders');
```

The product category which order is high is Clothing &Apparel 

```python
sales_amo = diwali_df.groupby(['Product_Category', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False).head(15)

sns.set(rc= {'figure.figsize':(30,5)})
sns.barplot(data = sales_amo , x='Product_Category', y= 'Amount', hue = 'Gender');
```

In Food more amount spend by the people

### Conclusion

Unmarried female spend more money than married female. Female spend more than male. Female spend more amount on food.
